```python
# import data to use and structure it
import pandas as pd
import numpy as np
df= pd.read_csv("https://raw.githubusercontent.com/regan-mu/ADS-April-2022/main/Assignments/Assignment%202/banking_churn.csv")
df.head()
```

```python
# prepare your data
# data information
df.info()
```

```python
#find missing values in the data
df.isna().sum()
# the data has no missing values
```

```python
# analyse the data types you are dealing with
df.dtypes
# noted object data types that we need to drop before making a model
```

```python
# drop the categorical data types and give our dataframe a new variable name
new_df = df.drop(['Surname','Geography','Gender','CustomerId','Age','RowNumber'], axis=1)
new_df
```

```python
# split our data into target variable and independent variables
# use the scikit learn alogarithm to sepatare train and test data
from sklearn.model_selection import train_test_split
X = new_df.drop(columns='CreditScore',axis=1)
y = new_df['CreditScore']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size =0.8,random_state=40)
X_train,X_test,y_train,y_test
```

```python
# next step is model building and score evaluation using sklearn
np.random.seed(42)
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
